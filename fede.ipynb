```python
import pickle
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('volumes_CA1_0.pkl', 'rb')
data = pickle.load(file)
file.close()
```

```python
plt.plot(np.log(data))
```

```python
def func(x, a, b, c):
    return a - b*1/(x**(c))  # a * (1- 1/(x**(b))) + c
    #return a * (1- np.exp(-b * x)) +c

def fit_func_and_plot(func, x_data, y_data,in_guess):
    popt, pcov = curve_fit(func, x_data, y_data, in_guess ,maxfev=50000,ftol=1.49012e-10, xtol=1.49012e-10)
    a, b, c = popt
    print(f"Fitted parameters: a={a}, b={b}, c={c}")
    plt.figure(figsize=(10, 6))
    plt.plot(x_data, y_data, 'bo', label='Data')  # Original data
    plt.plot(x_data, func(x_data, *popt), 'r-', label='Fitted function')  # Fitted curve
    plt.xlabel('Index')
    plt.ylabel('Volume')
    plt.legend()
    
    plt.show() 
```

```python
file = open('volumes_CA1_0.pkl', 'rb')
data = pickle.load(file)
file.close()

x_data = np.arange(1,len(data)+1)[2:]
y_data = np.log(np.array(data))[2:]

fit_func_and_plot(func=  func,x_data = x_data,y_data= y_data, in_guess = [y_data[-1] , y_data[-1] - y_data[0], 0.0001])


file = open('volumes_CA1_0.3.pkl', 'rb')
data = pickle.load(file)
file.close()

x_data = np.arange(1,len(data)+1)[2:]
y_data = np.log(np.array(data))[2:]

fit_func_and_plot(func,x_data,y_data,[y_data[-1] , y_data[-1] - y_data[0], 0.0001])


file = open('volumes_CA1_0.7.pkl', 'rb')
data = pickle.load(file)
file.close()

x_data = np.arange(1,len(data)+1)[2:]
y_data = np.log(np.array(data))[2:]

fit_func_and_plot(func,x_data,y_data,[y_data[-1] , y_data[-1] - y_data[0], 0.0001])
```
